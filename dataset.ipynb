```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.collections import EllipseCollection
from matplotlib.colors import Normalize

# Load the dataset
iris = pd.read_csv('data/iris.csv')

# Filter out non-numeric columns
numeric_iris = iris.select_dtypes(include=[np.number])

# Compute the correlation matrix
correlation_matrix = numeric_iris.corr()

# Define the function for plotting correlation ellipses
def plot_corr_ellipses(data, figsize=None, **kwargs):
    ''' Plot correlation ellipses to visualize correlation strength '''
    M = np.array(data)
    if not M.ndim == 2:
        raise ValueError('data must be a 2D array')
    fig, ax = plt.subplots(1, 1, figsize=figsize, subplot_kw={'aspect':'equal'})
    ax.set_xlim(-0.5, M.shape[1] - 0.5)
    ax.set_ylim(-0.5, M.shape[0] - 0.5)
    ax.invert_yaxis()

    # xy locations of each ellipse center
    xy = np.indices(M.shape)[::-1].reshape(2, -1).T

    # set the relative sizes of the major/minor axes according to the strength of
    # the positive/negative correlation
    w = np.ones_like(M).ravel() + 0.01
    h = 1 - np.abs(M).ravel() - 0.01
    a = 45 * np.sign(M).ravel()

    ec = EllipseCollection(widths=w, heights=h, angles=a, units='x', offsets=xy,
                           norm=Normalize(vmin=-1, vmax=1),
                           transOffset=ax.transData, array=M.ravel(), **kwargs)
    ax.add_collection(ec)

    # if data is a DataFrame, use the row/column names as tick labels
    if isinstance(data, pd.DataFrame):
        ax.set_xticks(np.arange(M.shape[1]))
        ax.set_xticklabels(data.columns, rotation=90)
        ax.set_yticks(np.arange(M.shape[0]))
        ax.set_yticklabels(data.index)

    return fig, ax, ec

# Plot correlation ellipses
fig, ax, ec = plot_corr_ellipses(correlation_matrix, figsize=(8, 6), cmap='bwr_r')

# Create a colorbar for the EllipseCollection
cb = plt.colorbar(ec, ax=ax)
cb.set_label('Correlation coefficient')

plt.title('Correlation Ellipses of Iris Dataset')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np

# Load the Iris dataset
iris = pd.read_csv('data/iris.csv')

# Convert PetalLengthCm into categorical bins
bins = [0, 2, 4, 6, 8]
labels = ['Short', 'Medium', 'Long', 'Very Long']
iris['PetalLengthCat'] = pd.cut(iris['PetalLengthCm'], bins=bins, labels=labels)

# Create a crosstabulation table
crosstab = pd.crosstab(index=iris['Species'], columns=iris['PetalLengthCat'], margins=True)
print("Crosstabulation Table:")
print(crosstab)

# Convert all columns to float for percentage calculation
crosstab = crosstab.astype(float)

# Calculate percentages
df = crosstab.copy()
df.loc[:, 'Short':'Very Long'] = df.loc[:, 'Short':'Very Long'].div(df['All'], axis=0)
df['All'] = df['All'] / sum(df['All'])

# Reset the 'All' column to a percentage representation
df['All'] = df['All'] * 100

# Rename index for clarity
df.index = df.index + ' (%)'

print("\nPercentage Crosstabulation Table:")
print(df)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = pd.read_csv('data/iris.csv')

# Define the hexbin function
def hexbin(x, y, color, **kwargs):
    cmap = sns.light_palette(color, as_cmap=True)
    plt.hexbin(x, y, gridsize=25, cmap=cmap, **kwargs)

# Create a FacetGrid for different species
g = sns.FacetGrid(iris, col='Species', col_wrap=2)
g.map(hexbin, 'SepalLengthCm', 'SepalWidthCm', 
      extent=[4, 8, 2, 4.5])
g.set_axis_labels('Sepal Length (cm)', 'Sepal Width (cm)')
g.set_titles('Species: {col_name}')

plt.tight_layout()
plt.show()
```
